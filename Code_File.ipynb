```python
# Add the Cost_USD column
df['Cost_USD'] = df['Revenue_USD'] * 0.6
print(df)
```

```python
# Add the Profit_USD column
df['Profit_USD'] = df['Revenue_USD'] - df['Cost_USD']
print(df)
```

```python
# Add the Profit_Margin column
df['Profit_Margin'] = (df['Profit_USD'] / df['Revenue_USD']) * 100
print(df)
```

```python
import numpy as np
customer_types = ["New", "Returning", "Wholesale", "Retail"]
marketing_channels = ["Email", "Social Media", "Direct", "Google Ads", "Referral"]
df["Customer_Type"] = np.random.choice(customer_types, size=len(df))
df["Marketing_Channel"] = np.random.choice(marketing_channels, size=len(df))
```
